```python
%%writefile app.py

from keybert import KeyBERT
import streamlit as st
import pandas as pd
import plotly.express as px

st.set_page_config(
    page_title="Response Analysis Dashboard",
    page_icon="🧑‍🎓",
    layout="wide",
    initial_sidebar_state="collapsed"
)

st.title("Response Analysis Dashboard")
st.markdown("### ")
st.markdown("*Please upload student data*")

uploaded_file = st.file_uploader("", type=["csv"])

if uploaded_file is not None:
    st.success("File uploaded successfully!")
    if uploaded_file.name.endswith(".csv"):
        df = pd.read_csv(uploaded_file)
        st.write("📄 Preview of uploaded CSV:")
        st.dataframe(df)

        st.markdown("### ")
        st.markdown("*Visualizations*")

        kw_model = KeyBERT(model='all-MiniLM-L6-v2')

        for col in df.columns:
            if col.startswith("Q"):
                question_text = df.at[0, col]
                responses = df[col].dropna().astype(str).tolist()
                full_text = " ".join(responses)

                keywords = kw_model.extract_keywords(
                    full_text,
                    keyphrase_ngram_range=(1, 2),
                    stop_words='english',
                    top_n=5
                )
                labels = [kw for kw, score in keywords]
                scores = [round(score * 100, 2) for kw, score in keywords]

                with st.expander(f"📝 {question_text}"):
                    st.markdown("**Top Response Themes:**")
                    fig = px.pie(
                        names=labels,
                        values=scores,
                        title="Keyword Theme Distribution",
                        hole=0.3
                    )
                    st.plotly_chart(fig, use_container_width=True)

                    st.markdown("**Extracted Keywords:**")
                    for label, score in zip(labels, scores):
                        st.write(f"• {label} ({score}%)")
```
